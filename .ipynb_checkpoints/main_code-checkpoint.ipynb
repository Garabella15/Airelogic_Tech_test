```python
import pandas as pd
from fuzzywuzzy import fuzz
import musicbrainzngs
import requests
import json
```

```python
musicbrainzngs.set_useragent("python-musicbrainzngs", "0.7.1", "https://github.com/Garabella15")
```

```python
def browse_release():
    artist = "f90e8b26-9e52-4669-a5c9-e28529c47894"
    limit = 100
    offset = 0
    recordings = []
    title_of_songs=[]
    page = 1
#     print("fetching page number %d.." % page)
    result = musicbrainzngs.browse_recordings(artist=artist, includes=["artist-credits"],
                    limit=limit)
    # print(result)
    page_recordings = result['recording-list']
    recordings += page_recordings
    # release-count is only available starting with musicbrainzngs 0.5
    if "recording-count" in result:
            count = result['recording-count']
            print("")
    while len(page_recordings) >= limit:
        offset += limit
        page += 1
#         print("fetching page number %d.." % page)
        result = musicbrainzngs.browse_recordings(artist=artist, includes=["artist-credits"],
                            limit=limit, offset=offset)
        page_recordings = result['recording-list']
        recordings += page_recordings
    return recordings
```

```python
df = pd.DataFrame(browse_release())
```

```python
tally = df[df['artist-credit-phrase'].str.startswith('Snoop Dogg')]
```

```python
tally
```

```python
gam=tally['title'].tolist()
```

```python
gam
```

```python
ray=set(gam)
ray
```

```python
elements = list(ray)

results = [[name, [], 0] for name in elements]

for (i, element) in enumerate(elements):
    for (j, choice) in enumerate(elements[i+1:]):
        if fuzz.ratio(element, choice) >= 90:
            results[i][2] += 1
            results[i][1].append(choice)
            results[j+i+1][2] += 1
            results[j+i+1][1].append(element)

data = pd.DataFrame(results, columns=['name', 'duplicates', 'duplicate_count'])
```
